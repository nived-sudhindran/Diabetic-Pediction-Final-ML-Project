From the above Outlier chart, we need to handle only **BMI** feature.

**GenHlth, MentHlth and PhysHlth** are scales which are important and cannot be removed considering it as Outliers.

The rest of them are 0's (denotes NO) and 1's (denotes YES)  - means **True Outliers** , and so not preferred to be removed.

```python
# Removing Outliers using IQR Method

def remove_outliers(df,columns):
  df_filtered = df.copy()
  for cols in columns:
    Q1 = df_filtered[cols].quantile(0.25)
    Q3 = df_filtered[cols].quantile(0.75)
    IQR = Q3 - Q1
    lower_wisker = Q1 - 1.5*IQR
    upper_wisker = Q3 + 1.5*IQR
    df_filtered = df_filtered[(df_filtered[cols] >= lower_wisker) & (df_filtered[cols] <= upper_wisker)]
  return df_filtered

#cols_with_outliers = ['CholCheck','BMI','Stroke','HeartDiseaseorAttack','Veggies','HvyAlcoholConsump','AnyHealthcare','NoDocbcCost','GenHlth','MentHlth','PhysHlth','DiffWalk']
#cols_with_outliers = ['BMI','MentHlth','PhysHlth','GenHlth']
cols_with_outliers = ['BMI']
df_without_outlier = remove_outliers(df,cols_with_outliers)
```

# 11. **Build the ML Model:**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import roc_curve, auc
```

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter = 1000),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Support Vector Machine': SVC(probability = False),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'AdaBoost': AdaBoostClassifier()
}
```

```python
# Train and Evaluate Models
results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    tr_acc = model.score(X_train, y_train)
    results.append({"Model": name, "Accuracy": acc, "Train Accuracy": tr_acc})

    print(f"\n{name} Performance:")
    print(f"Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    plt.figure(figsize=(5, 4))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap="Blues")
    plt.title(f"{name} - Confusion Matrix")
    #plt.xlabel("Predicted")
    #plt.ylabel("Actual")
    plt.show()


results_df = pd.DataFrame(results).sort_values(by="Accuracy", ascending=False)
print("\nModel Performance Summary:")
print(results_df)

```

**HYPERTUNING**

*From the above evaluation process, we see that **Random Forest** has the highest accuracy. So now lets try to increase its performance through Hypertuning its parameters*

```python
from sklearn.model_selection import GridSearchCV
param_grid={
    #'n_estimators': [int(x) for x in np.linspace(start=100, stop=105, num=6)],
    'n_estimators': [int(x) for x in range(100,150,3)],
    #'max_depth': [10, None],
    'max_depth': [None],
    'random_state': [42],
    'min_samples_split': [5, None],
    'bootstrap': [False, True],
    'warm_start': [False, True],
}

grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)
```

**Pipelining**

*My Dataset has only Numerical Data in it.* Also no need to perform any encoding as most of the data are in Binary valued, which is 0's and 1's

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.compose import ColumnTransformer
```

```python
# It is created using the parameters obtained by hypertuning
my_rf_model = RandomForestClassifier(
    bootstrap = True,
    max_depth = None,
    min_samples_split = 5,
    n_estimators = 139,
    random_state = 42,
    warm_start = False,
)

# selecting only the features which were selected using SelectKBest Algorithm
selected_features = ['HighBP', 'HighChol', 'CholCheck', 'BMI', 'HeartDiseaseorAttack','HvyAlcoholConsump',
                     'AnyHealthcare', 'NoDocbcCost', 'GenHlth','MentHlth', 'DiffWalk', 'Age']
preprocessor = ColumnTransformer([
    ('keep', 'passthrough', selected_features)],
                                 remainder='drop')

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', my_rf_model)
])

# I had no missing values in dataset and also no Scaling. Only feature selection was done.

# Train the pipeline
pipeline.fit(X_train, y_train)

# Predict and evaluate
y_pred = pipeline.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```
